# scikit-learn 中的 RBF 核
此处我们依旧使用sklearn中提供好的moon数据集，看一下对这个非线性样本，RBF核函数下的SVM处理的效果

## 加载数据集

```python
import numpy as np
import matplotlib.pyplot as plt
```

首先我们打印处我们的数据集，为了可视化我们仍然取两个维度

```python
from sklearn import datasets

X, y = datasets.make_moons(noise=0.15, random_state=666)

plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

## 构建pipeline

- 在下面我们会接触到一个新的调用方式，就是sklearn中给我提供的Pipeline，利用pipeline我们可以方便的减少代码量同时让机器学习的流程变得直观
- 我们在我们的pipline中首先进行标准化，然后我们实例化一个SVC，并且指定它的核函数为rbf以及它需要的一个超参数

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

def RBFKernelSVC(gamma):
    return Pipeline([
        ("std_scaler", StandardScaler()),
        ("svc", SVC(kernel="rbf", gamma=gamma))
    ])
```

## 超参数gamma=1的RBFKernelSVC

```python
svc = RBFKernelSVC(gamma=1)
svc.fit(X, y)
```

```python
def plot_decision_boundary(model, axis):
    
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1, 1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1, 1),
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
    
    plt.contourf(x0, x1, zz, cmap=custom_cmap)
```

我们来打印出当我们核函数的超参数为1的时候拟合的效果

```python
plot_decision_boundary(svc, axis=[-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

## 超参数gamma=100的RBFKernelSVC

我们再将它的超参数改成100，同学们再看一下效果

```python
svc_gamma100 = RBFKernelSVC(gamma=100)
svc_gamma100.fit(X, y)
```

```python
plot_decision_boundary(svc_gamma100, axis=[-1.5, 2.5, -1.0, 1.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

经过上面这么多的测试以后，我们大致的能够得到一个结果：我们的超参数越小模型越简单，超参数越大模型越复杂，或者换一种说法，也就是当参数越小，模型越倾向于欠拟合，参数越大时，模型越倾向于过拟合
